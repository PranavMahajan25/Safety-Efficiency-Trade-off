```python
cue1_data = data_df[data_df['cue']==1]
cue2_data = data_df[data_df['cue']==2]
cue3_data = data_df[data_df['cue']==3]
cue4_data = data_df[data_df['cue']==4]
```

```python
n_subj = 4
cue1_actions = np.zeros((n_subj, 60))
cue2_actions = np.zeros((n_subj, 60))
cue3_actions = np.zeros((n_subj, 60))
cue4_actions = np.zeros((n_subj, 60))
for n in range(n_subj):
    cue1_actions[n, :] = cue1_data[cue1_data['participant']==n+1].action
    cue2_actions[n, :] = cue2_data[cue2_data['participant']==n+1].action
    cue3_actions[n, :] = cue3_data[cue3_data['participant']==n+1].action
    cue4_actions[n, :] = cue4_data[cue4_data['participant']==n+1].action
```

```python
cue1_actions_raster = []
cue2_actions_raster = []
cue3_actions_raster = []
cue4_actions_raster = []
for n in range(n_subj): # no-go/withdrawals will be rasters
    cue1_actions_raster.append(np.where(cue1_actions[n,:]==0)[0])
    cue2_actions_raster.append(np.where(cue2_actions[n,:]==0)[0])
    cue3_actions_raster.append(np.where(cue3_actions[n,:]==0)[0])
    cue4_actions_raster.append(np.where(cue4_actions[n,:]==0)[0])
```

```python
cue3_actions_raster
```

```python
lineoffsets1 = (0.5+np.arange(n_subj))/n_subj
linelengths1 = np.ones(n_subj)/n_subj
line_widths1 = 6

plt.eventplot(cue1_actions_raster, lineoffsets=lineoffsets1,
                    linelengths=linelengths1, linewidths=line_widths1, colors='lightgrey')
plt.plot(np.mean(cue1_actions, axis=0), 'k')
plt.title('Cue1 - Approach to avoid')
plt.xlabel('Trials')
plt.ylabel('Probability(approach)')
```

```python
lineoffsets1 = (0.5+np.arange(n_subj))/n_subj
linelengths1 = np.ones(n_subj)/n_subj
line_widths1 = 6

plt.eventplot(cue2_actions_raster, lineoffsets=lineoffsets1,
                    linelengths=linelengths1, linewidths=line_widths1, colors='lightgrey')
plt.plot(np.mean(cue2_actions, axis=0), 'k')
plt.title('Cue2 - Withdraw to avoid')
plt.xlabel('Trials')
plt.ylabel('Probability(approach)')
```

```python
lineoffsets1 = (0.5+np.arange(n_subj))/n_subj
linelengths1 = np.ones(n_subj)/n_subj
line_widths1 = 6

plt.eventplot(cue3_actions_raster, lineoffsets=lineoffsets1,
                    linelengths=linelengths1, linewidths=line_widths1, colors='lightgrey')
plt.plot(np.mean(cue3_actions, axis=0), 'k')
plt.title('Cue3 - Uncontrollable then Approach to avoid')
plt.xlabel('Trials')
plt.ylabel('Probability(approach)')
```

```python
lineoffsets1 = (0.5+np.arange(n_subj))/n_subj
linelengths1 = np.ones(n_subj)/n_subj
line_widths1 = 6

plt.eventplot(cue4_actions_raster, lineoffsets=lineoffsets1,
                    linelengths=linelengths1, linewidths=line_widths1, colors='lightgrey')
plt.plot(np.mean(cue4_actions, axis=0), 'k')
plt.title('Cue4 - Uncontrollable then Withdraw to avoid')
plt.xlabel('Trials')
plt.ylabel('Probability(approach)')
```
